```python
import pandas as pd
import numpy as np

from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, make_scorer, accuracy_score
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier


categorical_features = ['대분류']
numerical_features = ['총창업비용', '보증금', '교육비', '보험료']


preprocessor = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
    ('num', StandardScaler(), numerical_features)
])

X = df[categorical_features + numerical_features]
y = df['cluster']


models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42),
    'LightGBM': LGBMClassifier(random_state=42)
}


f1_scorer = make_scorer(f1_score, average='macro')
acc_scorer = make_scorer(accuracy_score)


cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


results = {}

for name, model in models.items():
    pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', model)
    ])
    
    f1_scores = cross_val_score(pipeline, X, y, cv=cv, scoring=f1_scorer)
    acc_scores = cross_val_score(pipeline, X, y, cv=cv, scoring=acc_scorer)
    
    results[name] = {
        'F1 Mean': np.mean(f1_scores),
        'Acc Mean': np.mean(acc_scores)
    }


print("모델별 성능 (5-Fold CV):\n")
for name, score in results.items():
    print(f"{name:20}: F1 = {score['F1 Mean']:.4f}, Accuracy = {score['Acc Mean']:.4f}")
```

```python
# 최종 모델 Gradient Boosting

model = GradientBoostingClassifier(random_state=42) 
preprocessor = ColumnTransformer(transformers=[
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
    ('num', StandardScaler(), numerical_features)
])

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', model)
])

pipeline.fit(X, y)  



user_input = pd.DataFrame([{
    '대분류': '외식',           # 사용자 입력 
    '총창업비용': 50000000,     # 사용자 입력 
    '보증금': 10000000,         # 사용자 입력
    '교육비': 3000000,
    '보험료': 1000000
}])


predicted_cluster = pipeline.predict(user_input)[0]
predicted_proba = pipeline.predict_proba(user_input)


print(f"예측된 클러스터 라벨(사업 유형): {predicted_cluster}")
print(f"클러스터별 예측 확률: {predicted_proba}")
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# 1. 클러스터 예측 
predicted_cluster = pipeline.predict(user_input)[0]
predicted_proba = pipeline.predict_proba(user_input)

print(f"예측된 클러스터 라벨(사업 유형): {predicted_cluster}")
print(f"클러스터별 예측 확률: {predicted_proba}")

# 2. 추천용 지표 생성: 연평균매출액, 총창업비용, 계약해지수 + 상호변경수 사용
recommend_df = df.copy()
recommend_df["불안정지표"] = recommend_df["계약해지수"] + recommend_df["상호변경수"]

cluster_df = recommend_df[recommend_df["cluster"] == predicted_cluster].copy()

# 3. 스케일링
score_cols = ["연평균매출액", "총창업비용", "불안정지표"]
scaler = MinMaxScaler()
scaled = scaler.fit_transform(cluster_df[score_cols])
scaled_df = pd.DataFrame(scaled, columns=[col + "_스케일" for col in score_cols])
cluster_df = pd.concat([cluster_df.reset_index(drop=True), scaled_df], axis=1)

# 4. 지표 방향 조정 및 가중치 적용
cluster_df["수익성_스케일"] = cluster_df["연평균매출액_스케일"]            # 높을수록 좋음
cluster_df["비용_스케일"] = 1 - cluster_df["총창업비용_스케일"]          # 낮을수록 좋음
cluster_df["안정성_스케일"] = 1 - cluster_df["불안정지표_스케일"]        # 낮을수록 좋음

# 가중치 
w1, w2, w3 = 0.4, 0.3, 0.3
cluster_df["종합점수"] = (
    cluster_df["수익성_스케일"] * w1 +
    cluster_df["비용_스케일"] * w2 +
    cluster_df["안정성_스케일"] * w3
)

# 5. 상위 브랜드 추천
top_brands = cluster_df.sort_values("종합점수", ascending=False)[
    ["브랜드명", "법인명", "연평균매출액", "총창업비용", "계약해지수", "상호변경수", "종합점수"]
].head(5)

print("\n 추천 브랜드 (예측된 클러스터 내 상위 5개):")
print(top_brands)
```
