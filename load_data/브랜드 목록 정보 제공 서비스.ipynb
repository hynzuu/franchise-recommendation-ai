### 가맹정보_브랜드 목록 정보 제공 서비스



```python
import requests
import pandas as pd
import time

url = "http://apis.data.go.kr/1130000/FftcBrandRlsInfo2_Service/getBrandinfo"
service_key = "SERVICE_KEY" 

# 총 건수 및 반복 계산
total_count = 7501
rows_per_page = 100
total_pages = (total_count // rows_per_page) + 1

all_data = []

for page in range(1, total_pages + 1):
    params = {
        'serviceKey': service_key,
        'pageNo': page,
        'numOfRows': rows_per_page,
        'resultType': 'json',
        'jngBizCrtraYr': 2020
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        try:
            data = response.json()
            items = data.get("items", [])
            if items:
                all_data.extend(items)
                print(f"{page}/{total_pages} 페이지 완료")
            else:
                print(f"{page} 페이지: 데이터 없음")
        except Exception as e:
            print(f"페이지 {page} 파싱 실패:", e)
    else:
        print(f"{page} 페이지 요청 실패: {response.status_code}")
    
    time.sleep(0.5)

df = pd.DataFrame(all_data)
df.to_csv("brand_info_2020_full.csv", index=False, encoding="utf-8-sig")
print("CSV 저장 완료")
```
