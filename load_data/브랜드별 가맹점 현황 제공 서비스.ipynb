### 공정거래위원회_가맹정보_브랜드별 가맹점 현황 제공 서비스(2020)

- brand_frcs_stats_2020_full.csv로 저장


```python
# {
#   "resultCode": "00",
#   "resultMsg": "NORMAL SERVICE",
#   "numOfRows": "100",
#   "pageNo": "1",
#   "totalCount": 6338,
```

```python
import requests
import pandas as pd
import time

url = "http://apis.data.go.kr/1130000/FftcBrandFrcsStatsService/getBrandFrcsStats"
service_key =  "SERVICE_KEY" 

# 전체 페이지 수 계산
total_count = 6338
rows_per_page = 100
total_pages = (total_count // rows_per_page) + 1

all_data = []

for page in range(1, total_pages + 1):
    params = {
        'serviceKey': service_key,
        'pageNo': page,
        'numOfRows': rows_per_page,
        'resultType': 'json',
        'yr': 2020
    }

    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        items = data.get("items", [])
        if items:
            all_data.extend(items)
            print(f"{page}/{total_pages} 페이지 완료")
        else:
            print(f"{page} 페이지: 데이터 없음")
    else:
        print(f"{page} 페이지 요청 실패: {response.status_code}")
    
    time.sleep(0.5)  # 과도한 요청 방지

df = pd.DataFrame(all_data)
df.to_csv("brand_frcs_stats_2020_full.csv", index=False, encoding="utf-8-sig")
print("전체 CSV 저장 완료!")
```
