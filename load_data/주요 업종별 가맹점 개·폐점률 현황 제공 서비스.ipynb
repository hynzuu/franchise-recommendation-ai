
### 주요 도소매별 가맹점 개폐점 현황 조회

```python
import requests
import pandas as pd

# API 설정
url = "http://apis.data.go.kr/1130000/FftcIndutyFrcsOpclStatsService/getIndutyFrcsOpclWhrtStats"
service_key =  "SERVICE_KEY" 
year = 2021

params = {
    "serviceKey": service_key,
    "pageNo": 1,
    "numOfRows": 100,
    "resultType": "json",
    "yr": year
}

response = requests.get(url, params=params)

if response.status_code == 200:
    try:
        data = response.json()
        items = data.get("items", [])
        df = pd.DataFrame(items)
        print(df.head())

        # 저장
        df.to_csv(f"업종별_가맹점_개폐업통계_{year}.csv", index=False, encoding="utf-8-sig")
        print("CSV 저장 완료")
    except Exception as e:
        print("파싱 오류:", e)
else:
    print(f"요청 실패: {response.status_code}")
```

```python
import requests
import pandas as pd
import time

url = "http://apis.data.go.kr/1130000/FftcIndutyFrcsOpclStatsService/getIndutyFrcsOpclWhrtStats"
service_key =  "SERVICE_KEY" 

years = [2020, 2021, 2022, 2023, 2024]
all_data = []

for year in years:
    params = {
        "serviceKey": service_key,
        "pageNo": 1,
        "numOfRows": 100,
        "resultType": "json",
        "yr": year
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        try:
            data = response.json()
            items = data.get("items", [])
            for item in items:
                item["yr"] = year  # 연도 정보 추가
                all_data.append(item)
            print(f"{year}년 데이터 수집 완료 ({len(items)}건)")
        except Exception as e:
            print(f"{year}년 파싱 오류:", e)
    else:
        print(f"{year}년 요청 실패: {response.status_code}")
    
    time.sleep(0.3)

df = pd.DataFrame(all_data)
df.to_csv("업종별_가맹점_개폐업통계_2020_2024.csv", index=False, encoding="utf-8-sig")
print("CSV 저장 완료")
```
