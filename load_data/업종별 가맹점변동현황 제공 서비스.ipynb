### 업종별 가맹점 변동현황 

```python
import requests
import pandas as pd
import time

url = "http://apis.data.go.kr/1130000/FftcIndutyFrcsStatsService/getIndutyFrcsStatsService"
service_key =  "SERVICE_KEY" 


years = [2020, 2021, 2022, 2023, 2024]
all_data = []

for year in years:
    params = {
        "serviceKey": service_key,
        "pageNo": 1,
        "numOfRows": 100,
        "resultType": "json",
        "yr": year
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        try:
            data = response.json()
            items = data.get("items", [])
            for item in items:
                item["yr"] = year
                all_data.append(item)
            print(f"{year}년 데이터 수집 완료 ({len(items)}건)")
        except Exception as e:
            print(f"{year}년 파싱 오류:", e)
    else:
        print(f"{year}년 요청 실패: {response.status_code}")

    time.sleep(0.3)

df = pd.DataFrame(all_data)
df.to_csv("업종별_가맹점_변동현황_2020_2024.csv", index=False, encoding="utf-8-sig")
print("CSV 저장 완료")
```
